# Finding what roles each champion was played in

```python
import pandas as pd

file_path = './data/merged_data.csv'
df = pd.read_csv(file_path)

champion_roles = {}

for index, row in df.iterrows():
    champion = row['champion']
    role = row['position']

    if champion in champion_roles:
        if role not in champion_roles[champion]:
            champion_roles[champion].append(role)
    else:
        champion_roles[champion] = [role]

output_path = './data/champion_roles.txt'
with open(output_path, 'w') as f:
    for champion, roles in champion_roles.items():
        f.write(f"{champion}: {', '.join(roles)}\n")
```

# Calculate Teamscore Pre Random Forest

```python
import pandas as pd

file_path = 'data/pre_random_forest.csv'
ranking_table = pd.read_csv(file_path)
ranking_df = pd.DataFrame(ranking_table)

role_dict = {}
for _, row in ranking_df.iterrows():
    player = row.name
    teamname = row['Team']
    score = row['Score with region multiplier']
    if teamname not in role_dict:
        role_dict[teamname] = []
    role_dict[teamname].append(score)

team_score_no_role = {}
for team in role_dict:
    team_score_no_role[team] = sum(role_dict[team]) / len(role_dict[team])

sorted_team_score_no_role = sorted(team_score_no_role.items(), key=lambda x: x[1], reverse=True)

output_file = "Team_Ranking_pre_random_forest.txt"
with open(output_file, "w") as file:
    num = 1
    file.write("Team scores ranking without Role Weights:")
    for team, score in sorted_team_score_no_role:
        file.write(f"\n{num}. {team}: {score:.3f}")
        num += 1
```

# Calculate Teamscore Post Random Forest

```python
import pandas as pd

role_score = {"Top":0.251362, "Support": 0.072163, "Jungle":0.385088, "Mid":0.181071, "Bot": 0.110315}

file_path = 'data/post_random_forest.csv'
ranking_table = pd.read_csv(file_path)
ranking_df = pd.DataFrame(ranking_table)

role_dict = {}
for _, row in ranking_df.iterrows():
    player = row.name
    teamname = row['Team']
    score = row['Score']
    if teamname not in role_dict:
        role_dict[teamname] = []
    role_dict[teamname].append(score)

team_score = {}
for team in role_dict:
    team_score[team] = sum(role_dict[team]) / len(role_dict[team])

sorted_team_score = sorted(team_score.items(), key=lambda x: x[1], reverse=True)

output_file = "Team_ranking_post_random_forest.txt"
with open(output_file, "w") as file:
    file.write("Team scores ranking with Role Weights:")
    num = 1
    for team, score in sorted_team_score:
        file.write(f"\n{num}. {team}: {score:.3f}")
        num += 1
    num = 1
```
